```python
# NOTE that the file bpy.txt is some random geometry not equilibrum... for now

#usually you will need to import some packages, as the base python doesn't have everything
#How do you know? usually by searching google.
#Basic python can do the very basic stuff, memory assignment, simple arithmetic, read/write input/output.
import numpy as np
import fileinput
import random

for i in range(5):
    num = ('1.inp','2.inp','3.inp','4.inp','5.inp')
    
    bpy = './bpy.txt'
    string1 = ("H     1.0     3.30724     3.08352     0.00108","Cl    17.0     3.86084     3.09446     0.69776")
    string2 = ("replace","hi")

# for now let us focus on doing this combination just at two positions...

    new_file=open(num[i],mode="w",encoding="utf-8")
    for line in fileinput.FileInput(bpy,inplace=0):
        if 'H     1.0     3.30724     3.08352     0.00108' in line:
            choice = random.choice(string1)
            line = line.rstrip()
            line = line.replace(line, choice + '\n')
        if 'H     1.0    -2.40459     0.41535    -0.00697' in line:
            choice = random.choice(string2)
            line = line.rstrip()
            line = line.replace(line, choice + '\n')
        print(line, end='')
        new_file.write(line)

new_file.close()

    
```
