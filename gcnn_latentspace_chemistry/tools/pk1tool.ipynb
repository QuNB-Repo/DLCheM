This tool scans the extracted O embeddings for only one embedding per molecule (that of the most acidic oxygen, depending on an acidic order list), important for pk1 regression

```python
import numpy as np
import pandas as pd
```

```python
embs_path = 'data/datasets/embspKA-stef/601Oembslab.csv'

embs = pd.read_csv(embs_path,delimiter=',')

ldalabels = embs['ldalabel'].tolist()
gnumarkers = embs['gnumarker'].tolist()
gnucolors = embs['gnucolor'].tolist()

# Create a dictionary to map ldalabels to (gnucolor, gnumarker) tuples
ldalabel_to_gnu_dict = {}

for i, ldalabel in enumerate(ldalabels):
    gnumarker = gnumarkers[i]
    gnucolor = gnucolors[i]
    ldalabel_to_gnu_dict[ldalabel] = [gnucolor, gnumarker]

print(ldalabel_to_gnu_dict)
```

use dictionary to map ldalabels of 601O found with stefano to their gnucolors and gnumarkers


```python
stef_pk1_path = 'data/datasets/embspKA-stef/601embs.csv'

stef_pk1_data = pd.read_csv(stef_pk1_path,delimiter=',')

pk1_ldalabels = stef_pk1_data['ldalabel']

print(len(pk1_ldalabels))

# Create lists to store the mapped gnucolor and gnumarker values
gnucolors_mapped = []
gnumarkers_mapped = []

# Loop through the incomplete_ldalabels list and map the values using the dictionary
for ldalabel in pk1_ldalabels:
    if ldalabel in ldalabel_to_gnu_dict:
        gnucolor, gnumarker = ldalabel_to_gnu_dict[ldalabel]
        gnucolors_mapped.append(gnucolor)
        gnumarkers_mapped.append(gnumarker)
    else:
        pass

print(gnucolors_mapped)
print(gnumarkers_mapped)

print(len(gnucolors_mapped))

np.savetxt('colors.csv',gnucolors_mapped,delimiter=',')
np.savetxt('markers.csv',gnumarkers_mapped,delimiter=',')
```
