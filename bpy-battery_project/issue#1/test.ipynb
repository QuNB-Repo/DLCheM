```python
import numpy as np
import fileinput
import random

for i in range(5):
    num = ('test1.inp','test2.inp','test3.inp','test4.inp','test5.inp')
    
    bpy = '1.inp'
    string1 = ('MAXIT=30 ICHARG=2 MULT=1 ISPHER=1 $END')
    string2 = (' $STATPT HESS=READ $END' + '\n' ' $PCM SOLVNT=DMF SMD=.TRUE. $END' + '\n' ' $FORCE METHOD=SEMINUM $END')

    new_file=open(num[i],mode="w",encoding="utf-8")
    for line in fileinput.FileInput(bpy,inplace=0):
        if 'MAXIT=30 ICHARG=0 MULT=1 ISPHER=1 $END' in line:
            choice = random.choice(string1)
            line = line.rstrip()
            line = line.replace(line, choice +'\n')
        if ' $STATPT HESS=READ $END''H     1.0    -2.69741     0.60890    -0.60830' in line:
            choice = random.choice(string2)
            line = line.rstrip()
            line = line.replace(line, choice + '\n')
        print(line, end='')
        new_file.write(line)
    
new_file.close()
```
