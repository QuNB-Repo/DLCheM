```python
import torch
import schnetpack as spk
import math
from schnetpack.datasets import QM9


import numpy as np
from numpy import savetxt
```

```python
def hydrogen_nbh_matrix(number_atoms,atom_numb,x,y,z):
    #Returns the connections to the hydrogens
    count_H = 0
    for i in range(number_atoms):
        if atom_numb[i] == 1:
            count_H = count_H + 1
            
    hydrogen_nbh_distx = np.zeros((count_H,number_atoms))
    hydrogen_nbh_disty = np.zeros((count_H,number_atoms))
    hydrogen_nbh_distz = np.zeros((count_H,number_atoms))
    hydrogen_nbh_distance = np.zeros((count_H,number_atoms))
    
    count_H=0
    for i in range(number_atoms):
        if atom_numb[i] == 1:
            for j in range(number_atoms):
                hydrogen_nbh_distx[count_H][j] = x[i] - x[j]
                hydrogen_nbh_disty[count_H][j] = y[i] - y[j]
                hydrogen_nbh_distz[count_H][j] = z[i] - z[j]
                hydrogen_nbh_distance[count_H][j] = math.sqrt(hydrogen_nbh_distx[count_H][j]**2 + hydrogen_nbh_distx[count_H][j]**2 + hydrogen_nbh_distx[count_H][j]**2)
            count_H=count_H+1
    
    nbh_idx=np.zeros((count_H))
    for i in range(count_H):
        nbh_next_shortest = hydrogen_nbh_distance[i][0] 
        nbh_next_minimum = 0
        for j in range(number_atoms):
            if nbh_next_minimum < nbh_next_shortest < hydrogen_nbh_distance[i][j] and atom_numb[j]!=1:
                nbh_next_shortest = hydrogen_nbh_distance[i][j]
                nbh_next_minimum = hydrogen_nbh_distance[i][j]
                nbh_idx[i] = j
            
    print(hydrogen_nbh_distance)
    return nbh_idx
            


    
def Neighbors_Regression(num_mol):
    
    from schnetpack.datasets import QM9

    qm9data = QM9('./qm9.db', download=True, remove_uncharacterized=True)

    for idx in range(num_mol):
        
        ##Loading important data, converting tensors to numpy arrays 
        at, props = qm9data.get_properties(idx)
        x = props['_positions'][ :,0]
        y = props['_positions'][ :,1]
        z = props['_positions'][ :,2] 
    
        x=x.detach().numpy()
        y=y.detach().numpy()
        z=z.detach().numpy()
        
        number_atoms = len(z)
        
        atom_numb = props['_atomic_numbers']
        
        atom_numb = atom_numb.detach().numpy()
        
        nbh_idx = hydrogen_nbh_matrix(number_atoms,atom_numb,x,y,z)
        
        
        
        
        
        

```
