```python
from schnetpack.datasets import QM9

import numpy as np
from numpy import genfromtxt, savetxt
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
```

```python
embfilepath = '../../../data/main_data/embeddings/5000/dataH-distributions.csv'

emb = pd.read_csv(embfilepath,delimiter=',')


bin_position = -10.0
bin_length = 0.1
number_bins = 200

data = np.zeros((number_bins,2))

label = np.zeros((number_bins))
for j in range(number_bins):
    for i in range(len(emb)):
        if (bin_position-0.01) < emb['emb2'][i] < (bin_position):
            data[j][0] = data[j][0] + 1
            data[j][1] = bin_position
            label[j] = emb['Target'][i]
    bin_position = bin_position + bin_length
        

print(data)

    
color=['black','purple','orange','gold','pink',
        'green','skyblue','blue','brown','lime',
        'cyan','grey','darkslategrey','tan','lightsteelblue',
        'navy','violet','cornflowerblue','lightsalmon','palegreen',
        'sandybrown','peru','magenta','tomato','chocolate',
        'khaki','wheat','indigo','teal','salmon','olive']


ax = plt.axes()
x = np.linspace(-3,3,200)
#ax.set_xlabel('2D Distance')
ax.scatter(x,data[:,0],c=label,cmap=matplotlib.colors.ListedColormap(color))
```
