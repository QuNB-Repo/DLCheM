```python
import pandas as pd
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits import mplot3d
import matplotlib.patches as mpatches

import numpy as np
# read excel (examples of how to set up your data given in folder)
data = pd.read_csv("../../../../data/regression_charges/regression-U0pca.csv") 


df = pd.DataFrame(data,columns=['emb1', 'emb2','emb3','emb4','emb5','emb6','emb7','emb8','emb9','emb10',
                                'emb11','emb12','emb13','emb14','emb15','emb16','emb17','emb18','emb19','emb20',
                                'emb21','emb22','emb23','emb24','emb25','emb26','emb27','emb28','emb29','emb30',
                                'charges'])

print(df)
```

```python
X = df[['emb1', 'emb2','emb3','emb4','emb5','emb6','emb7','emb8','emb9','emb10',
                                'emb11','emb12','emb13','emb14','emb15','emb16','emb17','emb18','emb19','emb20',
                                'emb21','emb22','emb23','emb24','emb25','emb26','emb27','emb28','emb29','emb30']]
Y = df[['charges']]

#print(df[['emb1']])

ax = plt.axes(projection='3d')
plt.figure(figsize=(20,20))
ax.set_xlabel('First Neighbor')
ax.set_ylabel('Second Neighbor')
ax.set_zlabel('Atomization Energy')


ax.scatter(df[['emb1']],df[['emb2']],df[['charges']])

```
