```python
import pandas as pd
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits import mplot3d
import matplotlib.patches as mpatches

import numpy as np
# read excel (examples of how to set up your data given in folder)
data = pd.read_csv("../../../../data/neighbors_regression/dist_wgt/neighbor1.csv") 


df = pd.DataFrame(data,columns=['1st','2nd','3rd','4th','5th','6th','7th','AE'])

print(df)
```

```python
X = df[['1st','2nd','3rd','4th','5th','6th','7th']]
Y = df[['AE']]

#print(df[['emb1']])

ax = plt.axes(projection='3d')
plt.figure(figsize=(20,20))
ax.set_xlabel('First Neighbor')
ax.set_ylabel('Second Neighbor')
ax.set_zlabel('Atomization Energy')


ax.scatter(df[['1st']],df[['2nd']],df[['AE']])

```

```python
reg = LinearRegression()
reg.fit(X,Y)

reg.score(X,Y)
```

```python
reg.intercept_

print(reg.intercept_)

reg.coef_

y_hat = reg.predict(X)

print(y_hat)
print(reg.intercept_[0])
print(reg.coef_)


x = np.linspace(-2500,-500,10)
y = np.linspace(-2500,-500,10)
x1,y1 = np.meshgrid(x,y)
Z = -0.000701341098*x1 + 0.0000787098983*y1 - 18.08703254465



ax = plt.axes()
plt.figure(figsize=(20,20))
ax.set_xlabel('First Neighbor')
ax.set_ylabel('Atomization Energy')
ax.scatter(df[['1st']],df[['AE']])
ax.plot(df[['1st']],y_hat,'r-')
```
