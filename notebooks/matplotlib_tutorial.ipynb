```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

%reload_ext autoreload
```

```python
x1 = np.linspace(0,5,10)
y1 = x1**2
y2 = x1**4
plt.plot(x1,y1)
plt.title('x1 squared')
plt.xlabel('x1')
plt.ylabel('y1')
```

```python
plt.subplot(2,1,1)
plt.plot(x1,y1,'r')
plt.xlabel('x1')
plt.ylabel('y1')

```

```python
plt.subplot(2,1,2)
plt.plot(x1,y2,'b')
plt.xlabel('x1')
```

```python
fig_1 = plt.figure(figsize=(5,4),dpi=100)
ax = fig_1.add_axes([0.1,0.1,0.9,0.9])
ax.set_xlabel('x1')
ax.set_ylabel('y1')
ax.set_title('x1 Squared Plot')
ax.plot(x1,y1,label='plot 1')
ax.plot(y1,x1,label='plot 2')
ax.legend(loc=0)
ax2 = fig_1.add_axes([0.5,0.5,0.3,0.3])
ax2.set_xlabel('x1')
ax2.set_ylabel('y1')
ax2.set_title('x1 Squared Plot')
ax2.plot(x1,y1,label='plot 1')
ax2.plot(y1,x1,label='plot 2')
ax2.legend(loc=0)
ax2.text(0,35,'message')
```

```python
fig_3, ax3 = plt.subplots(figsize=(8,4),nrows=1,ncols=3)
plt.tight_layout()
ax3[1].set_title('Plot 2')
ax3[1].set_xlabel('x')
ax3[1].set_ylabel('y')
ax3[1].plot(x1,y1)
```

```python
fig_4 = plt.figure(figsize=(6,4))
ax4 = fig_4.add_axes([0,0,1,1])
ax4.plot(x1,y1,color='navy',alpha=0.75,lw=2,
         ls='-.',marker='o',markersize=7,
         markerfacecolor='y',markeredgewidth=0.5)
ax4.set_xlim([0,3])
ax4.set_ylim([0,8])
ax4.grid(True,color='0.6',dashes=(5,10,1,2))
ax4.set_facecolor('#FAEBD7')
```
